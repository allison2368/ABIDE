```python
import numpy as np
import pandas as pd
df = pd.read_csv("abide_behavioral.csv")
df.columns 
```

```python
new_df = df[['PIQ', 'VIQ', 'FIQ']]
# replace missing values with 0
for col in new_df.columns:
    new_df[col] = new_df[col].replace(np.nan, 0)

small_df = df['ADOS_GOTHAM_SEVERITY']
dfdf = small_df.replace(np.nan, 0)

ados_list = []
#print(df['ADOS_GOTHAM_SEVERITY'])
for i in dfdf:
    ados_list.append(i)

print(ados_list)
ind_first = ados_list.index(1.0)
ind_last = ados_list.index(0)
print(ind_first, ind_last)


newnew_df = pd.concat([new_df, small_df])
```

```python
data = df[['PIQ', 'VIQ', 'FIQ', 'ADOS_GOTHAM_SEVERITY', 'AGE_AT_SCAN']]

# remove nan values 
for col in data.columns:
    data[col] = data[col].replace(np.nan, 0)
short_df = (data[44:317]) 
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate the covariance matrix
covariance_matrix = data.cov()
print(covariance_matrix)

# Set up the matplotlib figure
plt.figure(figsize=(10, 8))

# Plot the covariance matrix as a heatmap
sns.heatmap(covariance_matrix, cmap='coolwarm', annot=False, fmt=".2f")

# Add title and rotate the x-axis labels for better readability
plt.title('Covariance Matrix')
plt.xticks(rotation=45)
plt.yticks(rotation=0)

# Show plot
plt.show()
```
